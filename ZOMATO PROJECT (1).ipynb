**LOAD DATASET**

```python
import pandas as pd
import os
import pandas as pd

if os.path.exists("/content"):
    # Running in Google Colab
    zomato_df = pd.read_csv("/content/sample_data/zomato.csv", encoding="latin-1")
    country_df = pd.read_excel("/content/sample_data/Country-Code.xlsx")
else:
    # Running locally / GitHub
    df = pd.read_csv("data/zomato.csv", encoding="latin-1")
    country_df = pd.read_excel("data/Country-Code.xlsx")

```

**TO MERGE BOTH THE DATASET**

```python
df = pd.merge(
    zomato_df,
    country_df,
    on="Country Code",
    how="left"
)
```

```python
df.head()
```

**REORGANIZE COLUMN INDEX**

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()

```

**Shows percentage share of restaurants by country**

```python
import matplotlib.pyplot as plt

country_names = df.country.value_counts().index
country_val = df.country.value_counts().values

plt.figure(figsize=(6,6))  # optional: make it square
plt.pie(
    country_val[:3],
    labels=country_names[:3],
    autopct='%1.2f%%'
)
plt.title("Top 3 Countries by Number of Restaurants")
plt.show()
```

**ANALYSIS BASED ON RESTAURANT**

```python
restaurant_analysis = df.groupby("restaurant_name")[["has_table_booking","is_delivering_now"]] \
                     .apply(lambda x: (x == "Yes").sum()) \
                     .reset_index()

restaurant_analysis
```

**ONLINE DELIVERY AVAILABILITY BY COUNTRY**

```python
country_analysis = df.groupby("country")[["has_online_delivery","is_delivering_now"]] \
                     .apply(lambda x: (x == "Yes").sum()) \
                     .reset_index()

country_analysis
```

```python
df[df['has_online_delivery']=='Yes'].groupby('country').size()
```

```python
import matplotlib.pyplot as plt

# Count online delivery = Yes by country
online_delivery_by_country = (
    df.groupby("country")["has_online_delivery"]
    .apply(lambda x: (x == "Yes").sum())
    .sort_values(ascending=False)
    .head()
)

# Plot
plt.figure(figsize=(10,5))
plt.bar(online_delivery_by_country.index, online_delivery_by_country.values)
plt.title("Top Countries Offering Online Delivery")
plt.xlabel("Country")
plt.ylabel("Number of Restaurants with Online Delivery")
plt.xticks(rotation=45)
plt.show()
```

**ONLINE DELIVERY IMPACT ON RATING**

```python
df.groupby("has_online_delivery")["aggregate_rating"].mean()
```

```python
import matplotlib.pyplot as plt

# Calculate mean rating by online delivery
delivery_rating = df.groupby("has_online_delivery")["aggregate_rating"].mean()

# Plot
plt.figure()
plt.bar(delivery_rating.index, delivery_rating.values)
plt.title("Average Rating by Online Delivery Availability")
plt.xlabel("Has Online Delivery")
plt.ylabel("Average Rating")
plt.show()
```

**FEATURE ENGINEERING**


**CREATE PRICE CATEGORY**

```python
def price_category(cost):
    if cost <= 500:
        return "Budget"
    elif cost <= 1500:
        return "Mid-range"
    else:
        return "Premium"

df["price_category"] = df["average_cost_for_two"].apply(price_category)
```

**PRICE CATEGORY IMPACT ON VOTES AND RATING**

```python
price_analysis = df.groupby("price_category")[["votes", "aggregate_rating"]].mean().reset_index()
price_analysis
```

**AVERAGE RATING BY COUNTRY**

```python
country_rating = df.groupby("country")["aggregate_rating"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(country_rating.index, country_rating.values)
plt.title("Top 10 Countries by Average Restaurant Rating")
plt.xlabel("Country")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()




```
